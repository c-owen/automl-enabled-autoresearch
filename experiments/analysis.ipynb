# autoresearch — session analysis

Reads a session's logs + git and renders the search trajectory. Point it at a session via `LOGS_DIR` (default `logs`) and the repo via `REPO_DIR` (default `.`). To re-render a **finished/archived** run with correct keep/discard, set `HEAD_REF` to its branch (e.g. `autoresearch/20260601-073738-...`) and `LOGS_DIR` to its archived logs. Keep/discard is from git ancestry; intent from commit messages.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from tools.extract_decisions import extract_decisions

LOGS_DIR = os.environ.get('LOGS_DIR', 'logs')
REPO_DIR = os.environ.get('REPO_DIR', '.')
HEAD_REF = os.environ.get('HEAD_REF', 'HEAD')   # set to a run's branch to re-render an old run
FIGDIR = os.path.join(os.path.dirname(os.path.normpath(LOGS_DIR)) or '.', 'figures')
os.makedirs(FIGDIR, exist_ok=True)

df = extract_decisions(LOGS_DIR, repo_dir=REPO_DIR, head=HEAD_REF)
print(f'Loaded {len(df)} trials from {LOGS_DIR}')
df.head()
```

## Summary counts

```python
print('n trials   :', len(df))
print('by status  :', df['status'].value_counts().to_dict())
print('by family  :', df['model_family'].value_counts().to_dict())
print('kept (git) :', int((df['kept_on_branch'] == True).sum()))
print('discarded  :', int((df['kept_on_branch'] == False).sum()))
print('crashes    :', int((df['status'] == 'crash').sum()))
```

## Progress curve

Trial index vs running-best `val_logloss`; red dashed lines mark family swaps.

```python
valid = df[df['status'] != 'crash'].copy()
valid['running_best'] = valid['val_logloss'].cummin()
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(valid['trial_id'], valid['val_logloss'], 'o-', alpha=0.5, label='val_logloss')
ax.plot(valid['trial_id'], valid['running_best'], 's-', color='green', label='running best')
for tid in df.loc[df['family_changed_from_prior'] == True, 'trial_id']:
    ax.axvline(tid, color='red', ls='--', alpha=0.3)
ax.set_xlabel('trial'); ax.set_ylabel('val_logloss')
ax.set_title('Progress curve (red dashed = family swap)'); ax.legend()
fig.savefig(os.path.join(FIGDIR, 'progress_curve.png'), dpi=120, bbox_inches='tight')
plt.show()
```

## Family distribution

```python
fig, ax = plt.subplots(figsize=(6, 4))
df['model_family'].value_counts().plot.bar(ax=ax, title='Trials per family')
ax.set_ylabel('count')
fig.savefig(os.path.join(FIGDIR, 'family_distribution.png'), dpi=120, bbox_inches='tight')
plt.show()
```

## Locus of change

```python
fig, ax = plt.subplots(figsize=(6, 4))
df['locus_of_change'].value_counts(dropna=False).plot.bar(ax=ax, title='Locus of change')
ax.set_ylabel('count')
fig.savefig(os.path.join(FIGDIR, 'locus_distribution.png'), dpi=120, bbox_inches='tight')
plt.show()
```

## Kept vs discarded per family

Color = git-derived outcome: green kept, orange discarded, red crash.

```python
def _color(row):
    if row['status'] == 'crash':
        return 'red'
    if row['kept_on_branch'] == True:
        return 'green'
    if row['kept_on_branch'] == False:
        return 'orange'
    return 'gray'

families = list(df['model_family'].dropna().unique())
fam_idx = {f: i for i, f in enumerate(families)}
fig, ax = plt.subplots(figsize=(9, 4))
for _, r in df.iterrows():
    ax.scatter(fam_idx.get(r['model_family'], 0), r['val_logloss'],
               c=_color(r), s=60, alpha=0.7)
ax.set_xticks(range(len(families))); ax.set_xticklabels(families, rotation=30)
ax.set_ylabel('val_logloss')
ax.set_title('Kept (green) / discarded (orange) / crash (red)')
fig.savefig(os.path.join(FIGDIR, 'kept_vs_discarded.png'), dpi=120, bbox_inches='tight')
plt.show()
```

## AUBC — area under the best-so-far curve (protocol §6.1)

Mean of the best-so-far `val_logloss` over trials 1..N (lower is better; rewards finding good solutions early). Counts **all** trials by `trial_id` — agent and BO-episode alike, since BO trials draw from the same budget.

```python
from tools.metrics import aubc, best_so_far

ordered = df.sort_values('trial_id')
losses = list(ordered['val_logloss'])
valid = [l for l in losses if pd.notna(l)]
print(f'trials                 : {len(losses)}')
print(f'best val_logloss       : {min(valid) if valid else float("nan"):.6f}')
print(f'AUBC (mean best-so-far): {aubc(losses):.6f}')
```

## Best-so-far curve (all trials)

Running-best `val_logloss` over every trial by index. BO-episode trials are shaded so you can see where the agent spent budget on sealed search.

```python
ordered = df.sort_values('trial_id').reset_index(drop=True)
bsf = best_so_far(list(ordered['val_logloss']))
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(ordered['trial_id'], bsf, 'o-', color='purple', label='best-so-far')
# Shade BO-episode trials, if any, to show where budget went to sealed search.
if 'source' in ordered.columns and (ordered['source'] == 'bo').any():
    for tid in ordered.loc[ordered['source'] == 'bo', 'trial_id']:
        ax.axvspan(tid - 0.5, tid + 0.5, color='orange', alpha=0.15)
ax.set_xlabel('trial'); ax.set_ylabel('best-so-far val_logloss')
ax.set_title('Best-so-far over trials (orange band = BO-episode trial)')
ax.legend()
fig.savefig(os.path.join(FIGDIR, 'best_so_far.png'), dpi=120, bbox_inches='tight')
plt.show()
```

## Tool engagement (C1 only — protocol §6.3)

BO uptake vs. utility: how often the agent opened an episode, how much budget it spent there, how often it adopted an episode best, and what share of session incumbents (best-so-far improvements) came from BO. On a C0 session this cell reports nothing — there is no tool.

```python
if 'source' in df.columns and (df['source'] == 'bo').any():
    bo = df[df['source'] == 'bo']
    n_episodes = bo['bo_episode_id'].nunique()
    bo_trials = len(bo)
    total_trials = len(df)
    n_adoptions = int(df['adopted_from_episode'].sum())

    # Incumbent attribution: which trials set a new best-so-far, by source.
    ordered = df.sort_values('trial_id')
    best = float('inf'); inc_bo = 0; inc_total = 0
    for _, r in ordered.iterrows():
        v = r['val_logloss']
        if pd.notna(v) and v < best:
            best = v; inc_total += 1
            if r['source'] == 'bo':
                inc_bo += 1

    print(f'episodes (invocations)   : {n_episodes}')
    print(f'episode family           : {bo.groupby("bo_episode_id")["model_family"].first().to_dict()}')
    print(f'budget per episode       : {bo["bo_episode_id"].value_counts().sort_index().to_dict()}')
    print(f'trial share to BO        : {bo_trials}/{total_trials} = {bo_trials / total_trials:.1%}')
    print(f'adoptions (episode-best) : {n_adoptions}  (rate {n_adoptions / n_episodes:.2f} per episode)')
    print(f'incumbents from BO       : {inc_bo}/{inc_total}')
else:
    print('C0 session (no BO capability) — no tool engagement to report.')
```
